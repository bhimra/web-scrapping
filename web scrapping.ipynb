```python
import requests
```



Now we will use requests to make a GET HTTP request for the following url =  'https://www.python.org/jobs/' by making a GET request.

GET - Used to request data from a specified resource.

POST - Used to send data to a server to create/update a resource.


```python
url = 'https://www.python.org/jobs/'
req = requests.get(url)
```

```python
req
```

```python
req.text[:200]
```

This shows that we have now grabbed the HTML of this web page. So now we can use Beautiful Soup (bs4) module to parse he HTML and retrieve the event data.

```python
#pip3 install bs4
from bs4 import BeautifulSoup
```

Let's create a BeautifulSoup object and pass it the HTML text.

```python
soup =BeautifulSoup(req.text,'lxml')
```

```python
soup
```

```python
events = soup.find('ol',{'class':'list-recent-jobs'})
```

```python
events
```

```python
type(events.findAll())
```

```python
events.findAll('li')
```

```python
for event in events.findAll('li'):
    d = dict()
    d['name'] = event.find('span',{'class':'listing-company-name'}).find('a').text
    d['time'] = event.find('span',{'class':'listing-posted'}).find('time').text
    d['location'] = event.find('span', {'class':'listing-location'}).text
    print(d)
```

```python
import pandas as pd
df = pd.DataFrame()

df['name'] = [e.find('a').text for e in events.findAll('span',{'class':'listing-company-name'})]
df['time'] = [e.find('time').text for e in events.findAll('span',{'class':'listing-posted'})]
df['location'] = [e.find('span', {'class':'listing-location'}).text for e in events.findAll('li')]
```
